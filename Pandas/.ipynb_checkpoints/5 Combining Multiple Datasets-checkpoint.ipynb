```python
import pandas as pd
import numpy as np

unemp_county = pd.read_csv('../datasets/output.csv')
unemp_county.head()
```

```python
df = pd.read_csv('../datasets/minwage.csv')

act_min_wage = pd.pivot_table(df, values="Department.Of.Labor.Cleaned.Low.Value.2020.Dollars", index="Year", columns='State')
act_min_wage = act_min_wage.replace(0, np.NaN).dropna(axis=1)
act_min_wage.head()
```

```python
def get_min_wage(year, state):
    try:
        return act_min_wage.loc[year][state]
    except:
        return np.NaN

get_min_wage(2012, 'Colorado')
```

```python
%%time

unemp_county['min_wage'] = tuple(map(get_min_wage, unemp_county['Year'], unemp_county['State']))
unemp_county.head()
```

```python
unemp_county[['Rate', 'min_wage']].corr()
```

```python
unemp_county[['Rate', 'min_wage']].cov()
```

```python
pres16 = pd.read_csv('../datasets/pres16results.csv')
pres16.head()
```

```python
county_2015 = unemp_county.copy()[(unemp_county['Year'] == 2015) & (unemp_county['Month'] == 'February')]
county_2015.head()
```

```python
pres16['st'].unique()
```

```python
state_abbv = pd.read_csv('../datasets/state_abbv.csv', index_col=0)
state_abbv = state_abbv[['Postal Code']]
state_abbv.head()
```

```python
state_abbv_dict = state_abbv.to_dict()['Postal Code']
state_abbv_dict
```

```python
county_2015['State'] = county_2015['State'].map(state_abbv_dict)
county_2015.tail()
```
